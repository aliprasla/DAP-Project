```python
#Number of crimes committed in each district per hour
crime['count']=1
Dist_hour=pd.pivot_table(crime,
                         columns=['Hour of the Day'],
                         values='count',
                         index=['District'],
                         aggfunc=sum)
Dist_hour
```
